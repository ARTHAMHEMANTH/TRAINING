```python
def common_elements(list1, list2):
    result = []
    for i in list1:
        if i in list2 and i not in result:
            result.append(i)
    return result
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
print(common_elements(list1, list2)) 
```

```python
def word_count(sentence):
    words = sentence.split()
    count = {}
    for w in words:
        if w in count:
            count[w] += 1
        else:
            count[w] = 1
    return count
sentence = "the cat and the hat"
print(word_count(sentence))
```
